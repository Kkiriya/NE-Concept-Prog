````python
def calculatrice(a:int, b:int) -> str:
    #operation
    add = a + b
    sub = a - b
    mult = a * b
    div = a / b
    moy = round((add + sub + mult + div)/ 4, 1)

    return f"""
    ===CALCULATRICE===
    {a} + {b} = {add}
    {a} - {b} = {sub}
    {a} * {b} = {mult}
    {a} / {b} = {div}
    Moyenne: {moy}
    """

print(calculatrice(5, 3))
````

````python
class CompteBancaire:
    nbr_compte = 0

    def __init__(self, nom:str, solde:float)-> None:
        self.nom = nom
        self.solde = solde
        CompteBancaire.nbr_compte += 1
        print(f"Création de compte: {nom}, solde: {solde}")
    
    def depot(self, montant:float)-> None:
        self.solde += montant
        print(f"Dépôt: {montant}, nouveau solde: {self.solde}")
    
    def retrait(self, montant:float)-> None:
        if montant > self.solde:
            print(f"Retrait impossible: {montant} demandé, solde: {self.solde}")
        else:
            self.solde -= montant
            print(f"Retrait: {montant}, nouveau solde: {self.solde}")
    
    @classmethod
    def afficher_nbr_compte(cls):
        print(f"Nombre de comptes créés: {cls.nbr_compte}")


alice = CompteBancaire("Alice", 100)
alice.depot(50)
alice.retrait(30)
alice.retrait(200)

b = CompteBancaire("Bob", 1)

CompteBancaire.afficher_nbr_compte()

````

### Exercice 3.1 : Filtre Personnalisé

```python
"""
Créez une fonction filtre_complexe qui:
1. Prend une liste de nombres ET une liste de conditions
2. Chaque condition est un tuple (opérateur, valeur)

4. Opérateurs supportés: >, <, >=, <=, ==, % (pair si %2==0)

Exemple:
nombres = [5, 12, 15, 20, 25, 30, 35]

Résultat: [12, 20] (car >10 ET <30 ET pair)

```

````python
nombres = [5, 12, 15, 20, 25, 30, 35]
conditions = [(">", 10), ("<", 30), ("%", 2)]

def filtre_complexe(ls_nombre, ls_condition):
    print(f"nombre = {ls_nombre}")
    print(f"condition = {ls_condition}")

    # contient le resultat de chaque filtre individuel pour ensuite ils prendre lintersection
    resultats_des_filtres = []
    for c in ls_condition:
        match(c[0]):
            case '>':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n > c[1]:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)

            case '<':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n < c[1]:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)

            case '>=':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n >= c[1]:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)

            case '<=':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n <= c[1]:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)

            case '==':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n == c[1]:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)

            case '%':
                # contien les resultat de ce filtre seulement
                resultat_filtre = set()
                for n in ls_nombre:
                    if n % 2 == 0:
                        # ajoute les nombres qui passe le filtre a la liste
                        resultat_filtre.add(n)
                # ajoute la liste a la liste qui contient tous les filtres
                resultats_des_filtres.append(resultat_filtre)
    
    resultat_final = set()
    for i in range(len(resultats_des_filtres)):
        if i == 0:
            resultat_final = resultats_des_filtres[i]
        else:
            resultat_final = resultat_final.intersection(resultats_des_filtres[i])

    resultat_str = f"resultat = {list(resultat_final)} (car " # contien le texte pour afficher le resultat
    for i, c in enumerate(ls_condition):
        if i < len(ls_condition) -1:
            resultat_str += f"{c[0]}{c[1]} AND "
        else:
            resultat_str += f"{c[0]}{c[1]})"

    print(resultat_str)

filtre_complexe(nombres, conditions)
````

````python
def filtre_complexe(nombres, conditions):
    resultat = []

    for nombre in nombres:
        valide = True

        for operateur, valeur in conditions:
            if operateur == ">":
                if not nombre > valeur:
                    valide = False
                    break
            elif operateur == "<":
                if not nombre < valeur:
                    valide = False
                    break
            elif operateur == ">=":
                if not nombre >= valeur:
                    valide = False
                    break
            elif operateur == "<=":
                if not nombre <= valeur:
                    valide = False
                    break
            elif operateur == "==":
                if not nombre == valeur:
                    valide = False
                    break
            elif operateur == "%":  # Pair
                if not nombre % 2 == 0:
                    valide = False
                    break

        if valide:
            resultat.append(nombre)

    return resultat

# Test
nombres = [5, 12, 15, 20, 25, 30, 35]
conditions = [(">", 10), ("<", 30), ("%", 2)]
print(filtre_complexe(nombres, conditions))  # [12, 20]
````

### Exercice 3.2 : Générateur de Profils

```python
"""
Créez un système qui génère des profils aléatoires:
1. Classe Personne: nom, age, profession, ville
2. Liste de noms prédéfinis: ["Alice", "Bob", "Charlie", "Diana"]
3. Liste de professions: ["Dev", "Designer", "Manager", "Testeur"]
4. Liste de villes: ["Montréal", "Toronto", "Vancouver", "Québec"]
5. Méthode generer_personne() qui crée une personne aléatoire
6. Méthode generer_groupe(n) qui crée n personnes uniques
7. Utilisez des sets pour éviter les doublons complets
"""
```

````python
import random

class Personne:

    def __init__(self, nom: str, profession: str, ville: str):
        self.nom = nom
        self.profession = profession
        self.ville = ville

    @staticmethod
    def generer_personne():
        noms = ["Alice", "Bob", "Charlie", "Diana"]
        professions = ["Dev", "Designer", "Manager", "Testeur"]
        villes = ["Montréal", "Toronto", "Vancouver", "Québec"]

        nom = random.choice(noms)
        profession = random.choice(professions)
        ville = random.choice(villes)

        return Personne(nom, profession, ville)

    @staticmethod
    def generer_groupe(n):
        groupe = set()
        for i in range(n):
            groupe.add(Personne.generer_personne())

        return groupe
    
    def __str__(self):
        return f"{self.nom}, {self.profession} à {self.ville}"
groupe = Personne.generer_groupe(3)
for p in groupe: 
    print(p)
````
