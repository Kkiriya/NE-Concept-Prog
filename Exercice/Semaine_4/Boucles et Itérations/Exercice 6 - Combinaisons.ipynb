### Exercice 6.1 : Statistiques d'une liste

Créez un programme qui calcule le minimum, maximum et moyenne d'une liste de nombres.

```python
import statistics

liste = [1, 5, 15]

print(liste)
print(f"minimum de la liste: {min(liste)}")
print(f"maximum de la liste: {max(liste)}")
print(f"moyenne de la liste: {sum(liste) / len(liste)}")
print(f"moyenne de la liste(mean): {statistics.mean(liste)}")
```

```python
#corriger car peut etre interessant
nombres = [12, 45, 8, 32, 19, 67, 23]

if nombres:
    minimum = nombres[0]
    maximum = nombres[0]
    somme = 0

    for nombre in nombres:
        if nombre < minimum:
            minimum = nombre
        if nombre > maximum:
            maximum = nombre
        somme += nombre

    moyenne = somme / len(nombres)

    print(f"Minimum : {minimum}")
    print(f"Maximum : {maximum}")
    print(f"Moyenne : {moyenne:.2f}")
else:
    print("La liste est vide")
```

### Exercice 6.2 : Palindrome

Écrivez un programme qui vérifie si un mot est un palindrome.

```python
palindrome = ["radar", "kayak", "level"]
mots = ["cheval", "pomme", "fromage"]

def est_Palindrome(mot):
    debut = 0
    fin = len(mot) - 1

    while debut < fin:
        if mot[debut] != mot[fin]:
            return False
        debut +=  1
        fin -= 1
    return True

for p in palindrome:
    print(f"{p} est un palindrome? \"{est_Palindrome(p)}\"")

for m in mots:
    print(f"{m} est un palindrome? \"{est_Palindrome(m)}\"")


print("\n\n")

#fonctionne aussi en utilisant le reversal en python mais prends plus despace dans la memoire
def est_Palindrome2(mot):
    if mot == mot[::-1]:
        return True
    else: 
        return False

for p in palindrome:
    print(f"{p} est un palindrome? \"{est_Palindrome2(p)}\"")

for m in mots:
    print(f"{m} est un palindrome? \"{est_Palindrome2(m)}\"")
```

### Exercice 6.3 : Jeu de devinette amélioré

Créez un jeu de devinette avec limite d'essais et indices.

```python
import random

nbr = random.randint(1,10)
diff = [3, 5, 9]
choix = diff[int(input("Veuillez choisir votre difficulté de: (1/2/3): ")) - 1]
nbr_essaies = choix

while(nbr_essaies > 0):
    inp = int(input(f"Veuillez devinez un nombre entre 1 et 10 ({nbr}) (nbr essair restant {nbr_essaies}/{choix}): ")) #le nombre ici est donner pour faciliter le testage

    if inp >= 1 and inp <= 10:
        if inp == nbr:
            print("Bravo! vous avez deviner le chiffre!")
            break
        else:
            nbr_essaies -= 1
            if inp < nbr: 
                print(f"Dommage!, {inp} est plus petit!")
            else:
                print(f"Dommage!, {inp} est plus grand!")
            
            if(nbr_essaies == 0):
                print("Dommage, vous avez perdu!")
    else:
        print(f"Erreur: Nombre invalide: {inp}")
        nbr_essaies -= 1
```
