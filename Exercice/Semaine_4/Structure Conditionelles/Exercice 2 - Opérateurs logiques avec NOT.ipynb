```python
#useful methods
def to_bool(a):
        true = ("true", "1", "yes", "y")
        false = ("false", "0", "no", 'n')

        a = a.strip().lower()

        if a in true:
            return True
        elif a in false:
            return False
```

### Exercice 2.1 : Accès autorisé
Créez un programme qui vérifie si l'accès est autorisé :
- L'utilisateur doit avoir un badge valide
- ET ne pas être banni
- ET avoir payé son abonnement


```python
badge_valide = to_bool((input("Badge valide? (y/n):")))
usr_banned = to_bool((input("usr banned? (y/n):")))
abb_paied = to_bool((input("abbonement payer? (y/n):")))

if badge_valide and (not usr_banned) and abb_paied:
    print("access autoriser")
else:
    print("access non autoriser")
```

### Exercice 2.2 : Conditions météorologiques
Écrivez un programme qui détermine si on peut sortir :
- Il ne doit pas pleuvoir
- ET la température doit être entre 15°C et 30°C
- ET il ne doit pas y avoir d'alerte météo


```python
rain = to_bool(input("pleut-t-il? (y/n):"))
temp = int(input("quel est la temperature?:"))
alert = to_bool(input("il y a-t-il une alerte meteo? (y/n):"))

if (not rain) and (temp >= 15 and temp <= 30) and (not alert):
    print("vous pouvez sortir")
else: 
    print("vous ne pouver pas sortir")
                    
```

### Exercice 2.3 : Validation de données
Créez un programme qui vérifie si des données sont valides :
- L'email ne doit pas être vide
- ET l'âge doit être entre 0 et 120
- ET le nom ne doit pas contenir de chiffres


```python
email = bool(input("veuillez entrez votre email:"))
age = int(input("veuillez entrez votre age:"))
nom = input("veuillez entrez votre nom:").isalpha()

if email:
    print("email est valide")
else:
    print("email est invalide")
    
if age > 0 and age < 120:
    print("age valide")
else:
    print("age est invalide")
    
if nom:
    print("nom valide")
else: 
    print("age est invalide")
```
