### 5.1 : Code à Développer

Créez une fonction `decrire_rectangle()` qui prend `longueur` et `largeur` comme paramètres positionnels et affiche les dimensions.

````python
def decrire_rectangle(longeur, largeur):
    print(f"longeur: {longeur}")
    print(f"largeur: {largeur}")
    print(f"perimetre: {(longeur + largeur) * 2}")
    print(f"aire: {longeur*largeur}")

decrire_rectangle(10,5)
````
