## Exercice 4 : Fonctions Natives et Ajout d'Éléments

### 4.1 : Fonctions de base

Pour `valeurs = [15, 2, 8, 20, 5]`, utilisez :

- `len()`, `max()`, `min()`, `sum()`
- `sorted()` sans modifier l'originale

```python
valeurs = [15, 2, 8, 20, 5]

print(len(valeurs))
print(max(valeurs))
print(min(valeurs))
print(sum(valeurs))
print(sorted(valeurs))
```

### 4.2 : Ajout d'éléments

Avec `liste = ["a", "b"]`, ajoutez des éléments en utilisant :

- `append()`
- `insert()`
- `extend()`

```python
liste = ["a", "b"]

liste.append('d')
print(liste)

liste.insert(2, 'c')
print(liste)

liste.extend(['e','f'])
print(liste)
```

### 4.3 : Conversion depuis string

Convertissez la string `"python"` en liste de caractères, puis `"1,2,3,4,5"` en liste d'entiers.


```python
str = "python"
nbr = "1,2,3,4,5"

liste1 = list(str)
print(liste1)

liste2 = [int(x) for x in nbr.split(',')]
print(liste2)
```
