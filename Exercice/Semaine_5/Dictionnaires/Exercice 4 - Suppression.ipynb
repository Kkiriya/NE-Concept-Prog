### 4.2 : Nettoyage conditionnel (Erreurs à corriger)

```python
# Ce code devrait supprimer toutes les valeurs None mais contient des erreurs
produits = {
    "livre": 15,
    "stylo": None,
    "cahier": 8,
    "gomme": None,
    "regle": 12
}

for cle in produits:
    if produits[cle] == None:
        del cle

print(produits)
```

````python
import copy

produits = {
    "livre": 15,
    "stylo": None,
    "cahier": 8,
    "gomme": None,
    "regle": 12
}

produits_tmp = copy.deepcopy(produits)

for cle in produits_tmp:
    if produits[cle] is None:
        del produits[cle]

print(produits)
````

````python
# Ce code devrait supprimer toutes les valeurs None mais contient des erreurs
produits = {
    "livre": 15,
    "stylo": None,
    "cahier": 8,
    "gomme": None,
    "regle": 12
}

cles_a_supprimer = []
for cle, valeur in produits.items():
    if valeur is None:
        cles_a_supprimer.append(cle)

for cle in cles_a_supprimer:
    if produits[cle] == None:
        del produits[cle]

print(produits)
````

````python
produits = {
    "livre": 15,
    "stylo": None,
    "cahier": 8,
    "gomme": None,
    "regle": 12
}

cles_a_supprimer = [cle for cle, valeur in produits.items() if valeur is None]

for cle in cles_a_supprimer:
    del produits[cle]

print(produits)
````

### 4.3 : pop() avec défaut (À compléter)

```python
config = {"theme": "sombre", "langue": "fr"}

# Supprimez 'theme' avec valeur par défaut "clair"
ancien_theme = ______
# Essayez de supprimer 'volume' avec défaut 50
volume = ______

print(f"Ancien theme: {ancien_theme}")

```

````python
config = {"theme": "sombre", "langue": "fr"}

# Supprimez 'theme' avec valeur par défaut "clair"
ancien_theme = config.pop('theme')
config.setdefault('theme', 'clair')
# Essayez de supprimer 'volume' avec défaut 50
config.setdefault('volume', '50')
volume = config.pop('volume')


print(f"Ancien theme: {ancien_theme}")
print(f"Volume: {volume}")
````

````python
config = {"theme": "sombre", "langue": "fr"}

ancien_theme = config.pop("theme", "clair")
volume = config.pop("volume", 50)

print(f"Ancien theme: {ancien_theme}")
print(f"Volume: {volume}")
````
