**lorsque l'on veut verifier une condition on utilise:**
    *if condition:*
        *code*

```python
if 4 > 2:
    print("4 est plus grand que deux")
```

**else permet d'executer du code dans le cas ou la condition n'est pas respecter**
    *else:*
        *code*

```python
if 2 > 4:
    print("4 est plus grand que deux")
else:
    print("4 n'est pas plus grand que deux")
```

**si on souhaite verifier un deuxieme condition on utilise:**
    *elif:*
        *code*

```python
if 2 > 2:
    print("4 est plus grand que deux")
elif 2 == 2:
    print("2 est egal a deux")
else:
    print("4 n'est pas plus grand que deux")
```
