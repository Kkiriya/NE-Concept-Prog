```python
age = 18

if age >= 18:
    adult = True
else: 
    adult = False

print(adult)
```

**Les operateur ternaire permette de transformer toute les ligne de code si dessus en une seule ligne de code**

```python
adult = True if age >= 18 else False
print(adult)
```

```python
# peut etre utilise avec tout type de variable
adult = 'est un adulte' if age >= 18 else 'est un enfant'
print(adult)
```

```python
#nous donne une erreur ici
age = None
adult = True if age >= 18 else False
```
