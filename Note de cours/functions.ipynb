# Les fonctions
De manières générale permet de reutiliser du code et avoir le meme comportement a plusieur endroits different

il faut utiliser le mot clés *def* pour creer une function puis on la nomme, 
Le naming convention pour les fonctions est le snake_case 

````python
def add(a,b):
    return a + b

print(add(3,2))
````

### Définitions des fonctions: 
Bien nommer les fonctions est important! Le nom devrait indique sa fonction

pour python declarer une fonctions fait la meme chose que declarer une variable, python stocke le code en memoire et y fait reference
```python
def le_nom_de_la_fonction():
    print("Hello World!")
```

````python
def le_nom_de_la_fonction():
    print("Hello World!")

le_nom_de_la_fonction()
````

### Retourner des valeurs:
pour retourner des valeurs on utilise return.

return renvois la valeur qui lui est donnees, donc lorsque l'on appelle la fonction elle correspond a la valeur du return

````python
def le_nom_de_la_fonction_2():
    return "Hello World!"

print(le_nom_de_la_fonction_2())
````

### Plusieurs return
python permet de faire plusieurs return, il le fait sous forme de tuple

````python
def multiple_retours():
    return 'Salut', 'Commment ca va?', 1, True

print(multiple_retours())
````

### Les parametres
permettent de donner des valeurs ou objet a la fonctions pour qu'elle puisse les utiliser
```python
def parametre(a, b)
```

````python
def addition_entre_deux_nombres(a, b):
    return a + b

print(addition_entre_deux_nombres(8,9))
````

### Valeur par default 
permettent deviter les erreurs si tous les argument ne sont pas fournit


````python
def addition_par_default(a,b = 2):
    return a + b

print(addition_par_default(3))
````

### Specifier les arguments: 
On peut specifier les arguments si on oublies l'ordre dans lequel il doivent etre ecris

````python
def greetings(nom, prenom):
    return f"Bonjour {prenom} {nom}"

print(greetings(prenom="Émile", nom="Valade"))
````

### Random stuff 

````python
a = 'Je suis a'
def my_func(b):
    print(b)

my_func(a)
````

````python
a = 1
def my_func(b):
    print(a is b)

my_func(a)
````

la variable a ne change pas ici car la function cree sa propre reference vers b donc a ne refere donc pas un nouvel objet 

de plus les string sont immuable

````python
a = 'Je suis a'
def my_func(b):
    b = 2
    print(b is a)

my_func(a)
print(a)
````

c change ici car c fait reference a un objet muable

````python
c = [1, 2]
def my_func_muable(b):
    b.append(3)
    print(c is b)

my_func_muable(c)
print(c)
````

````python
d = [5, 6]
def my_func_muable_2(b):
    b = [7]
    print(b is d)

my_func_muable_2(d)
print(d)
````

### Passer un nombre indéfini d'argument
permet de passer un nombre (quasi infini) d'argument 

recupere les valeurs et les mets dans un tuple

lorsque l'on utilise une etoile il faut specifier tout les argument qui ne doivent pas etre inclus

````python
def somme_de_tout(*args, factors): 
    print(args)
    print(sum(args))
    print(factors)

somme_de_tout(5, 47, 54, 98, factors="BOOM")
````

### Recuperer des arguemnts nommer non definis: 
les recupere et les mets dans un dict

````python
def recup_args(**kwargs):
    print(kwargs)

recup_args(nom="V", prenom="Émile", age=22)
````
