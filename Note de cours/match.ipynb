**match est l'equivalant d'un switch case**

```python
#demonstration avec un cascade de if else
error = 500

if error == 500: 
    print("Erreur serveur")
elif error == 400:
    print("Mauvaise requete")
elif error == 404:
    print("Ressource non trouvee")
else:
    print("erreur inconue")
```

```python
#demonstration avec match
match error:
    case 500:
        print("Erreur serveur")
    case 400:
        print("Mauvaise requete")
    case 404:
        print("Ressource non trouvee")
    case _:
        print("erreur inconue")
```

```python
#cas dexample avec des or
error = 500

if error == 500: 
    print("Erreur serveur")
elif error == 400:
    print("Mauvaise requete")
elif error == 404 or error == 406 or error == 405:
    print("Ressource non trouvee")
else:
    print("erreur inconue")
```

```python
#cas dexample avec des or
match error:
    case 500:
        print("Erreur serveur")
    case 400:
        print("Mauvaise requete")
    case 404 | 406 | 405:
        print("Ressource non trouvee")
    case _:
        print("erreur inconue")
```
